```python
##import all files
customer_df1 = pd.read_csv('data/file1.csv')
customer_df2 = pd.read_csv('data/file2.csv')
customer_df3 = pd.read_csv('data/file3.csv')
print(customer_df1.columns)
print(customer_df2.columns)
print(customer_df3.columns)
```

- Standardizing header names
- Deleting and rearranging columns – delete the column customer as it is only a unique identifier for each row of data
- Working with data types – Check the data types of all the columns and fix the incorrect ones (for ex. customer lifetime value and number of complaints )
- Filtering data and Correcting typos – Filter the data in state and gender column to standardize the texts in those columns
- Removing duplicates
- Replacing null values – Replace missing values with means of the column (for numerical columns)

```python
## define a function to make heading lowcase
def change_columns_headers(df):
    df.columns = df.columns.str.lower()
    return df
```

```python
# change all headers to lower case
change_columns_headers(customer_df3)
change_columns_headers(customer_df2)
change_columns_headers(customer_df1)

print(customer_df1.columns)
```

```python
# define a function that renames the columns to get the same naming
def change_column_headers(df, column_old_name, column_new_name):
    return df.rename(columns={column_old_name:column_new_name},inplace = True) #we are renaming the columns 
```

```python
change_column_headers(customer_df1,"st","state" )
customer_df1
change_column_headers(customer_df2,"st","state" )
customer_df2
```

```python
# Defines function that rearranges columns based on list to have the same order in all the files
def rearrange_columns(df, list_of_new_column_names):
    df = df[list_of_new_column_names]
    return df
```

```python
new_column_names = list(customer_df1.columns)
```

```python
new_column_names
```

```python
# Rearrange columns

customer_df2 = rearrange_columns(customer_df2,new_column_names)
customer_df3 = rearrange_columns(customer_df3,new_column_names) 
```

```python
# Use reset_index to not have any duplicate indexes. 
customer_df = pd.concat([customer_df3,customer_df1,customer_df2],axis = 0).reset_index(drop=True)
customer_df
```

```python
#defining function that drops columns
def drop_column(df,column_name):
    df.drop(columns=[column_name], inplace = True)
    return df
```

```python
drop_column(customer_df,'customer')
```

```python
customer_df.info()
```

```python
# There are 2060 values in "Costumer lifetime value column" with % at the end. This needs to be deleted
customer_df['customer lifetime value'].str.contains('%').count()
```

```python
# Delete the % at the end of the numbers
customer_df['customer lifetime value'] = customer_df['customer lifetime value'].replace({'%': ""}, regex=True)
```

```python
# Change date type to float and round.
customer_df['customer lifetime value'] = pd.to_numeric(customer_df['customer lifetime value'], errors='coerce')
customer_df['customer lifetime value'] = customer_df['customer lifetime value'].round()
```

```python
# Clean data in Number of open complains
customer_df['number of open complaints'].unique()
```

```python
# There are 2067 elements with /
customer_df['number of open complaints'].str.contains('/').sum()
```

```python
# Create a dictionary with all the replacements that need to be done
replacement_mapping_dic = {
    "1/0/00" : "0",
    "1/1/00" : "1",
    "1/2/00" : "2",
    "1/3/00" : "3",
    "1/4/00" : "4",
    "1/5/00" : "5",
    }

customer_df["number of open complaints"]=customer_df["number of open complaints"].replace(replacement_mapping_dic)
```

```python
# Elements with / are now deleted from the column
customer_df['number of open complaints'].unique()
```

```python
# Check unique elemens in the column 'State'
customer_df['state'].unique()
```

```python
customer_df["state"].value_counts()
```

```python
# Eliminate duplicate names for the States
customer_df["state"] = customer_df["state"].replace("WA", "Washington")
customer_df["state"] = customer_df["state"].replace("Cali", "California")
customer_df["state"] = customer_df["state"].replace("AZ", "Arizona")
```

```python
# If we take a look to the data now there are no duplicates
customer_df["state"].value_counts()
```

```python
# Check the values in 'Gender'
customer_df["gender"].value_counts()
```

```python
# Create a dictionary that serves as index for gender

customer_df["gender"] = customer_df["gender"].replace("Male", "M")
customer_df["gender"] = customer_df["gender"].replace("female", "F")
customer_df["gender"] = customer_df["gender"].replace("Femal", "F")
```

```python
# Now the data is not duplicated in Gender
customer_df["gender"].value_counts()
```

```python
# Removing duplicates
customer_df = customer_df.drop_duplicates()
```

```python
# If we print the table we can see that duplicated values were deleted
customer_df
```

```python
# Find out how many null values are in the dataset.
customer_df.isna().sum()
```

```python
customer_df['gender'] = customer_df['gender'].fillna('Unknown') ## fill the unknown as str
```

```python
# Now there are no more NA elements
customer_df['gender'].isna().sum()
```

```python
# Function to replaces the null values with the mean in this 3 columns
def replace_null_mean(df_columns):
    df_columns = df_columns.fillna(df_columns.mean())
    return df_columns
```

```python
replace_null_mean(customer_df['customer lifetime value'])
replace_null_mean(customer_df['monthly premium auto'])
replace_null_mean(customer_df['total claim amount'])
```
